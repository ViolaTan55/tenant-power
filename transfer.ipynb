```python
val_aql = "FOR x IN indiv2indiv RETURN x"
value_query_result = local_db.AQLQuery(val_aql,rawResults=True, batchSize = 1000)
col_value = {}
ind_val = 0

for value in value_query_result:
    col_value[ind_val] = value
    ind_val += 1
    
# create dataframe from dictionary of dictionaries
corp = pd.DataFrame.from_dict(data = col_value, orient = 'index')
indiv2indiv_ar = corp.to_numpy()
```

```python
def indiv2indiv(indiv2indiv_ar, batch_size):
    for i in range(0,2582):
        print(i)
        print(i*batch_size,(i+1)*batch_size)
        batch_ar = indiv2indiv_ar[i*batch_size:(i+1)*batch_size][:]

        for j in batch_ar:
            edge_attributes = {"_from": j[2], "_to": j[3], "_key":j[0],"type":j[5]}
            edge = server_db["indiv2indiv"].createDocument(edge_attributes)
            edge.save()
            
indiv2indiv(indiv2indiv_ar,1000)
```
